# Data Preparation

Before performing the analysis, both datasets need to be transformed into a common format. Since the trader dataset records individual trades with timestamps while the Fear & Greed dataset contains daily market sentiment, both datasets are aligned using a common **Date** column. This enables the comparison of trader behavior under different market sentiment conditions.

```python
historical_data["Timestamp"].head()
```

```python
fear_greed["timestamp"].head()
```

## Timestamp Conversion

The trader dataset and the Fear & Greed dataset store timestamps in Unix format. To enable date-based analysis, both timestamps are converted into a standard datetime format. A separate **Date** column is then extracted from each dataset, which will be used to merge the datasets at the daily level.

```python
# Convert historical data timestamp (milliseconds)
historical_data["Timestamp"] = pd.to_datetime(
    historical_data["Timestamp"],
    unit="ms"
)

# Extract date
historical_data["Date"] = historical_data["Timestamp"].dt.date


# Convert fear & greed timestamp (seconds)
fear_greed["timestamp"] = pd.to_datetime(
    fear_greed["timestamp"],
    unit="s"
)

# Convert date column
fear_greed["date"] = pd.to_datetime(fear_greed["date"])

# Extract date
fear_greed["Date"] = fear_greed["date"].dt.date
```

```python
historical_data[["Timestamp", "Date"]].head()
```

```python
fear_greed[["timestamp", "Date", "classification"]].head()
```

## Merging the Datasets

To analyze trader performance under different market conditions, the historical trading data is merged with the Fear & Greed Index using the common **Date** column. This enriches each trade with the corresponding market sentiment observed on that day.

```python
merged_data = historical_data.merge(
    fear_greed[["Date", "classification", "value"]],
    on="Date",
    how="left"
)

merged_data.head()
```

```python
merged_data.info()
```

```python
merged_data[["Date", "classification", "value"]].head()
```

# Feature Engineering

Feature engineering involves creating additional variables that provide deeper insights into trading behavior. These derived features simplify the analysis of trader performance across different market sentiment conditions.

The following metrics will be created:
- Trade Outcome (Profit or Loss)
- Absolute Profit/Loss
- Trade Size Category

```python
merged_data["Trade Outcome"] = np.where(
    merged_data["Closed PnL"] > 0,
    "Profit",
    "Loss"
)
```

```python
merged_data["Trade Outcome"].value_counts()
```

```python
merged_data["Absolute PnL"] = merged_data["Closed PnL"].abs()
```

```python
merged_data["Size USD"].describe()
```

```python
merged_data["Trade Size"] = pd.qcut(
    merged_data["Size USD"],
    q=4,
    labels=["Small", "Medium", "Large", "Very Large"]
)
```

```python
merged_data["Trade Size"].value_counts()
```

```python
merged_data.groupby("classification")["Closed PnL"].mean()
```

# Analysis 1: Average Profit/Loss by Market Sentiment

The first objective is to examine whether trader profitability changes under different market sentiment conditions. By calculating the average realized Profit and Loss (PnL) for each sentiment category, we can identify whether traders tend to perform better during periods of fear or greed.

```python
avg_pnl = (
    merged_data
    .groupby("classification")["Closed PnL"]
    .mean()
    .sort_values()
)

avg_pnl
```

```python
plt.figure(figsize=(8,5))

bars = plt.bar(
    avg_pnl.index,
    avg_pnl.values,
    color="skyblue",
    edgecolor="black"
)

plt.title("Average Closed PnL by Market Sentiment")
plt.xlabel("Market Sentiment")
plt.ylabel("Average Closed PnL (USD)")
plt.xticks(rotation=20)

plt.bar_label(bars, fmt="%.2f")

plt.tight_layout()
plt.show()
```

### Observation

- Traders generated the highest average profit during **Greed** market conditions.
- Average profitability declined during periods of **Fear** and **Extreme Fear**.
- This suggests that market sentiment may influence trading performance, with optimistic markets generally associated with better outcomes.

# Analysis 2: Win Rate by Market Sentiment

While average profit provides an overall performance measure, it may be influenced by a small number of exceptionally large trades. Therefore, win rate is calculated to determine the proportion of profitable trades under each market sentiment category.

```python
win_rate = (
    merged_data
    .groupby("classification")["Trade Outcome"]
    .apply(lambda x: (x == "Profit").mean() * 100)
    .sort_values()
)

win_rate
```

```python
plt.figure(figsize=(8,5))

bars = plt.bar(
    win_rate.index,
    win_rate.values,
    color="green",
    edgecolor="black"
)

plt.title("Win Rate by Market Sentiment")
plt.xlabel("Market Sentiment")
plt.ylabel("Win Rate (%)")
plt.xticks(rotation=20)

plt.bar_label(bars, fmt="%.1f%%")

plt.tight_layout()
plt.show()
```

### Observation

- Traders achieved the highest win rate during **Greed**.
- The lowest win rate was observed during **Extreme Fear**.
- The results indicate that trading success varies across different market sentiment conditions.

# Analysis 3: Trading Activity by Market Sentiment

Trading frequency is an important indicator of market participation. This analysis examines the number of trades executed under each market sentiment category to identify whether traders are more active during periods of fear or greed.

```python
trade_count = (
    merged_data["classification"]
    .value_counts()
    .sort_values()
)

trade_count
```

```python
plt.figure(figsize=(8,5))

bars = plt.bar(
    trade_count.index,
    trade_count.values,
    color="orange",
    edgecolor="black"
)

plt.title("Number of Trades by Market Sentiment")
plt.xlabel("Market Sentiment")
plt.ylabel("Number of Trades")

plt.bar_label(bars)

plt.tight_layout()
plt.show()
```

### Observation

- The highest trading activity was recorded during **Fear**, with **133,871 trades**, accounting for the majority of trading activity in the dataset.
- **Greed** ranked second with **36,289 trades**, while **Neutral** (7,141 trades) and **Extreme Greed** (6,962 trades) had significantly fewer trades.
- The results indicate that traders were substantially more active during fearful market conditions, suggesting increased market participation or volatility during periods of uncertainty.
- Since the number of trades varies considerably across sentiment categories, subsequent performance metrics should be interpreted while considering these differences in sample size.

# Analysis 4: Distribution of Profit/Loss by Market Sentiment

Average profit provides only a single summary value and may be influenced by a few exceptionally large trades. To better understand the consistency and spread of trader performance, a boxplot is used to visualize the distribution of Closed Profit and Loss across different market sentiment categories.

```python
plt.figure(figsize=(10,6))

sns.boxplot(
    data=merged_data,
    x="classification",
    y="Closed PnL",
    showfliers=False
)

plt.title("Distribution of Closed PnL by Market Sentiment")
plt.xlabel("Market Sentiment")
plt.ylabel("Closed PnL (USD)")

plt.tight_layout()
plt.show()
```

### Observation

- After removing the visualization of extreme outliers, the distribution of Closed PnL becomes easier to interpret across different market sentiment categories.
- **Extreme Greed** exhibits the widest interquartile range (IQR), indicating greater variability in trading outcomes compared to other market conditions.
- **Greed** also shows a relatively broad distribution with a higher upper quartile, suggesting the potential for larger profits during optimistic market conditions.
- **Neutral** market conditions display the narrowest spread, indicating more consistent but comparatively smaller trading outcomes.
- Overall, trader performance appears to be more volatile during periods of market optimism than during neutral market conditions.

# Analysis 5: Average Trade Size by Market Sentiment

Trade size reflects the amount of capital committed to each transaction. This analysis investigates whether traders allocate larger positions under different market sentiment conditions.

```python
avg_trade_size = (
    merged_data
    .groupby("classification")["Size USD"]
    .mean()
    .sort_values()
)

avg_trade_size
```

```python
plt.figure(figsize=(8,5))

bars = plt.bar(
    avg_trade_size.index,
    avg_trade_size.values,
    color="purple",
    edgecolor="black"
)

plt.title("Average Trade Size by Market Sentiment")
plt.xlabel("Market Sentiment")
plt.ylabel("Average Trade Size (USD)")

plt.bar_label(bars, fmt="%.0f")

plt.tight_layout()
plt.show()
```

# Analysis 6: Long vs Short Positions Across Market Sentiments

This analysis examines whether market sentiment influences the direction of trades by comparing the number of long and short positions executed under each sentiment category.

```python
position_dist = pd.crosstab(
    merged_data["classification"],
    merged_data["Direction"]
)

position_dist
```

```python
position_dist.plot(
    kind="bar",
    figsize=(9,5)
)

plt.title("Long vs Short Positions by Market Sentiment")
plt.xlabel("Market Sentiment")
plt.ylabel("Number of Trades")

plt.xticks(rotation=20)

plt.tight_layout()
plt.show()
```

# Analysis 8: Correlation Analysis

A correlation matrix is used to examine the relationships between key numerical trading variables and identify potential linear associations.

# Key Findings

Based on the exploratory analysis and the relationship between trader performance and the Fear & Greed Index, the following key observations were identified:

1. Market sentiment has a noticeable impact on trader performance.
2. The highest average realized profit was observed during **Greed** market conditions, indicating that traders generally performed better during optimistic market periods.
3. The highest trading activity occurred during **Fear**, suggesting increased market participation and volatility when investors were uncertain.
4. The win rate improved from **31.7%** during Neutral conditions to **49.0%** during Extreme Greed, indicating that profitable trades became more frequent during bullish market sentiment.
5. Profit and loss distributions contained several extreme outliers, demonstrating that a small number of trades contributed disproportionately to overall profitability.
6. After excluding outliers from visualization, trader performance appeared more consistent, although market sentiment continued to influence the spread of returns.

# Business Insights

The analysis provides several practical insights for traders and trading platforms:

- Trader behaviour changes noticeably with market sentiment.
- Fearful markets encourage significantly higher trading activity, possibly due to increased volatility and uncertainty.
- Although more trades occur during Fear, the highest average profitability was achieved during Greed.
- Higher win rates during Greed and Extreme Greed suggest that bullish market conditions may offer more favourable trading opportunities.
- Because trading performance contains significant outliers, evaluating traders solely on average profit may be misleading. Additional metrics such as win rate and profit consistency should also be considered.

# Recommendations

Based on the findings of this analysis, the following recommendations are proposed:

- Incorporate market sentiment as an additional feature in trading strategies and risk management models.
- Apply stricter risk controls during periods of Fear due to increased market volatility and trading activity.
- Evaluate trader performance using multiple metrics, including average profit, win rate, and risk-adjusted returns, rather than relying on a single measure.
- Continuously monitor sentiment indicators such as the Fear & Greed Index to support data-driven trading decisions.
- Future analyses could include additional market indicators, technical signals, and trader-specific behavioural features to improve predictive performance.
